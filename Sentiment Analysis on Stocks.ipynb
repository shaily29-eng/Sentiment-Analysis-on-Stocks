# Stock Sentiment Analysis using News Headlines

```python
import pandas as pd
```

```python
df=pd.read_csv('Combined_News_DJIA.csv',encoding="ISO-8859-1")#encoding use for this dataset as it will identift utf8 char in dataset and will read data
```

```python
df.head() #combining top 25 heading n applying nlp processing
#label column is dependent feature
```

```python
df.tail() 
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']#old data as train n new data as test as upcoming data will follow the previous data n act accordingly 
```

```python
# Removing punctuations
data=train.iloc[:,2:27]#reading all dataset and instead of ^a-zA-Z--replace with space 
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)#replacing spl char with spaces
#A regular expression is a sequence of characters that can be used as a search pattern.
#Sometimes referred to as Regex or Regexen
#shift+tab--press this before  ( to know the function property

# Renaming column names for ease of access-this is on input/independent feature/training dataset
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
#combining all texts from coln top 1 to top 25 and converting into vectors(or W2V:as word to vector can also be done)
' '.join(str(x) for x in data.iloc[1,0:25])
#iterating from column top 0 to 25 from row 1 to last([1,0:25]) and if row changes than there will be a space so we have perform join (' ')
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
#use of CountVectorisor or we can use Tfidf or word to vectors(W2V)..whichever used to get correct accuracy
# these are used for converting sentences into vectors and create bag of words
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier  #instead can use naivebais or even multi naive bais also
```

```python
#no need to remove stop words as this plays an imp role in news headlines to identify the unique feature as the cat, the dog ....
## implement BAG OF WORDS by CountVectorizer
#use of count vectorisor:eg: 3 sent suppose to convert into vector:1.the cat is running, 2.the dog is barking 3.the boy is singing
#now, it will combine and takeout the unique words from all sentence: the cat dog boy is running barking singing(these r all features):these all will be lined up on basis of frequency
#now,in first sentence:teh cat is running: so the unique words mark as 1(if present else 0) than predict y(result) as 1/0 based on all dataset 

countvector=CountVectorizer(ngram_range=(2,2))#ngram uses min and max parameters 
#ngram_range(2,2) means combining 2 words from sentence (don't remove stopwrds as news headlines me imp hota)
#here, eg:the cat (2 words combine) , the dog, the boy
#now, if these 2 words are present in sentence so use 1 else 0

traindataset=countvector.fit_transform(headlines)#using fit_transform in headlines is dataset(independent feature)
```

```python
traindataset[0]
#FYI: ans:sprase matrix uses 1 and 0  
```

```python
#instead of random forest can use hyper parametertuning--until here we are performing preprocessing in train set 
# implement RandomForest Classifier:in executing it takes time as it uses 200 decision trees
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label']) #fit(x data,y data)
```

```python
## Predict for the Test Dataset-- now this on test set
#same code used as train set used above:iteration on top 25 columns present and perform join function,count vectorizer,
#,prediction by random classifier and predicitng in test dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)  #gives classifictn matrix
print(matrix)
score=accuracy_score(test['Label'],predictions) #gives 84% accurancy score:here,good accurancy
print(score)
report=classification_report(test['Label'],predictions) #gives precision report 
print(report)
#hence,this is balanced dataset as it gives this 3 results(otherwise, imbalanced)
##################################################THE END##############################################################
```

```python
########repeating all the steps with use of Tifdivectorizer############### row 13
#TF-IDF is an abbreviation for Term Frequency-Inverse Document Frequency and
#,it is the most used algorithm to convert the text into vectors.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier 
```

```python
#implement bag of words created by TifdiVectorizer
tfidfvector=TfidfVectorizer(ngram_range=(2,2))
traindataset=tfidfvector.fit_transform(headlines)
```

```python
#FYI: r2_score :used for regression and accurancy_score:used for classification
```

```python
traindataset[0]     #it also uses sprax matrix as count vectorizer
```

```python
# implement RandomForest Classifier:in executing it takes time as it uses 200 decision trees
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label']) #fit(x data,y data)
```

```python
#predict for test dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = tfidfvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
matrix=confusion_matrix(test['Label'],predictions)  #gives classifictn matrix
print(matrix)
score=accuracy_score(test['Label'],predictions) #gives 85% accurancy score:here,good accurancy
print(score)
report=classification_report(test['Label'],predictions) #gives precision report 
print(report)
#hence,this is balanced dataset as it gives this 3 results(otherwise, imbalanced)
##################################################THE END##############################################################
```
